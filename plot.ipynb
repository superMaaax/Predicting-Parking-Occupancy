```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the original dataset
original_data = pd.read_csv("dataset.csv")
rate = original_data['rate'].values
cycle = original_data['cycle'].values
effect = original_data['indicator_norm'].values
# Convert OccupancyDateTime column to datetime and extract date strings only
original_data['OccupancyDateTime'] = pd.to_datetime(original_data['OccupancyDateTime'])
date_labels = original_data['OccupancyDateTime'].dt.strftime('%Y-%m-%d').values
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 9), sharex=True)

# RATE
axes[0].plot(rate, label='TREND', color='blue')
axes[0].set_ylabel('Rate')
axes[0].legend(loc='upper right')
axes[0].grid(True)

# CYCLE
axes[1].plot(cycle, label='CYCLE', color='red')
axes[1].set_ylabel('Cycle')
axes[1].legend(loc='upper right')
axes[1].grid(True)

# EFFECT
axes[2].plot(effect, label='EFFECT', color='orange')
axes[2].set_ylabel('Effect (norm)')
axes[2].legend(loc='upper right')
axes[2].grid(True)

# Shared X‑axis formatting
axes[2].set_xlabel('Date')
num_ticks = 15
total = len(date_labels)
ticks = np.linspace(0, total - 1, num=num_ticks, dtype=int)
axes[2].set_xticks(ticks)
axes[2].set_xticklabels(date_labels[ticks], rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure datetime
original_data['OccupancyDateTime'] = pd.to_datetime(
    original_data['OccupancyDateTime']
)

# Pick the date you want to inspect
date_str = '2025-03-22'
date_val = pd.to_datetime(date_str).date()

# Filter for that day
mask   = original_data['OccupancyDateTime'].dt.date == date_val
zoomed = original_data.loc[mask].reset_index(drop=True)

if zoomed.empty:
    print(f"No data for {date_str}.")
else:
    # Only plot TREND
    fig, ax = plt.subplots(figsize=(10, 4))
    
    ax.plot(
        zoomed['OccupancyDateTime'], 
        zoomed['rate'], 
        color='blue', 
        label='TREND'
    )
    ax.set_title(f"TREND on {date_str}")
    ax.set_ylabel("Rate")
    ax.set_xlabel("Time")
    ax.grid(True)

    fig.autofmt_xdate()
    plt.tight_layout()
    # plt.show()
    plt.savefig(f"trend_plot_{date_str}.png", dpi=300, bbox_inches='tight')
```
